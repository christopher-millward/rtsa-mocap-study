```python
# Import the data
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from utils.data_loading import load_participant_details
from utils.kinematics.general_helpers import create_rotation_matrices

participant_details = load_participant_details('../data/raw_normalized_data/participant_details.xlsx')
```

```python
# print the length of each participant's data in hours
lengths = []
for participant in participant_details:
    path = participant['filename']
    if participant['rtsa_side'] is None:
        continue
    data = np.loadtxt(f'../data/raw_normalized_data/{path}', delimiter='\t', skiprows=1, usecols=range(1,2))
    length = data.shape[0] / 10 / 2600
    lengths.append(
        {
            "participant": participant['filename'],
            "duration (hrs)": length
        }
    )

df = pd.DataFrame(lengths)
print(
    df.sort_values(by="duration (hrs)", ascending=False)
    .reset_index()
    .round(2)
    .to_markdown()
)
```
