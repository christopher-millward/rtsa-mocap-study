I found some old data that Dan sent me back in 2021. I remember there being a bunch of "dead" data at the end of the files, indicating that the participant had taken the shirt off, but never turned of the IMU systems. I just want to document how long these files were to make sure I wasn't just trimming data for the sake of doing something.

```python
import os 
import sys
import numpy as np
import pandas as pd
from utils.data_loading import load_participant_details

# set root folder to project root
root_path = os.path.abspath(os.path.join(".."))
if root_path not in sys.path:
    sys.path.insert(0, root_path)

# import the data
participant_details = load_participant_details('../raw_data/participant_details.xlsx')
```

```python
# print the length of each participant's data in hours
lengths = []
for participant in participant_details:
    path = participant['filename']
    data = np.loadtxt(f'../original_raw_data/{path}', delimiter='\t', skiprows=1, usecols=range(1,2))
    length = data.shape[0] / 10 / 2600
    lengths.append(
        {
            "participant": participant['filename'],
            "duration (hrs)": length
        }
    )

df = pd.DataFrame(lengths)
print(
    df.sort_values(by="duration (hrs)", ascending=False)
    .reset_index()
    .round(2)
    .to_markdown()
)
```
